```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 1
# What fraction of loans are healthy vs. problematic?
# Soudeep Ghosh
# ==========================================

import pandas as pd
import plotly.express as px   # Added for Plotly Charts

# Load Dataset
loan = pd.read_csv("completedloan.csv")

# ------------------------------------------
# Create Loan Category
#
# Healthy Loans:
# A = Contract Finished, No Problems
# C = Running Loan, OK
#
# Problematic Loans:
# B = Contract Finished, Loan Not Paid
# D = Running Loan, Client in Debt
# ------------------------------------------

loan["loan_category"] = loan["status"].map({
    "A": "Healthy",
    "B": "Problematic",
    "C": "Healthy",
    "D": "Problematic"
})

# ------------------------------------------
# Count Loans by Category
# ------------------------------------------

result = (
    loan.groupby("loan_category")
        .size()
        .reset_index(name="loan_count")
)

# ------------------------------------------
# Calculate Percentage
# ------------------------------------------

total_loans = result["loan_count"].sum()

result["percentage"] = (
    result["loan_count"] / total_loans * 100
).round(2)

# ------------------------------------------
# Sort Result
# ------------------------------------------

result = result.sort_values(
    by="loan_count",
    ascending=False
).reset_index(drop=True)

# ------------------------------------------
# Display Result
# ------------------------------------------

print("\nQuestion 1 Result")
print(result)

# ------------------------------------------
# Key Findings
# ------------------------------------------

healthy = result[result["loan_category"] == "Healthy"].iloc[0]
problematic = result[result["loan_category"] == "Problematic"].iloc[0]

print("\nKey Findings")

print(
    f"Healthy Loans: {healthy['loan_count']} "
    f"({healthy['percentage']}%)"
)

print(
    f"Problematic Loans: {problematic['loan_count']} "
    f"({problematic['percentage']}%)"
)

print(f"Total Loans: {total_loans}")

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.pie(
    result,
    names="loan_category",          # Slice labels
    values="loan_count",            # Slice values
    title="Healthy vs Problematic Loans",
    hole=0.4                        # Creates a Donut Chart
)

# Display Percentage and Label
fig.update_traces(
    textinfo="percent+label"
)

# Improve Layout
fig.update_layout(
    title_x=0.5,
    legend_title="Loan Category"
)

# Show Chart
fig.show()
```

```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 2
# What is the total number of loans by status
# (Active, Closed, and Defaulted)?
# Soudeep Ghosh
# ==========================================

import pandas as pd
import plotly.express as px

# Load Dataset
loan = pd.read_csv("completedloan.csv")

# ------------------------------------------
# Loan Status Classification
#
# Active:
# C = Running Loan, OK
#
# Closed:
# A = Contract Finished, No Problems
#
# Defaulted:
# B = Contract Finished, Loan Not Paid
# D = Running Loan, Client in Debt
# ------------------------------------------

# Create Loan Status Category
loan["loan_status"] = loan["status"].map({
    "A": "Closed",
    "B": "Defaulted",
    "C": "Active",
    "D": "Defaulted"
})

# ------------------------------------------
# Count Loans by Status
# ------------------------------------------

result = (
    loan.groupby("loan_status")
        .size()
        .reset_index(name="loan_count")
)

# ------------------------------------------
# Sort Result
# ------------------------------------------

status_order = ["Active", "Closed", "Defaulted"]

result["loan_status"] = pd.Categorical(
    result["loan_status"],
    categories=status_order,
    ordered=True
)

result = (
    result.sort_values("loan_status")
          .reset_index(drop=True)
)

# ------------------------------------------
# Display Result
# ------------------------------------------

print("\nQuestion 2 Result")
print(result)

# ------------------------------------------
# Key Findings
# ------------------------------------------

active = result[result["loan_status"] == "Active"].iloc[0]
closed = result[result["loan_status"] == "Closed"].iloc[0]
defaulted = result[result["loan_status"] == "Defaulted"].iloc[0]

print("\nKey Findings")

print(f"Active Loans: {active['loan_count']}")
print(f"Closed Loans: {closed['loan_count']}")
print(f"Defaulted Loans: {defaulted['loan_count']}")
print(f"Total Loans: {result['loan_count'].sum()}")

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.bar(
    result,
    x="loan_status",
    y="loan_count",
    text="loan_count",
    title="Total Number of Loans by Status",
    labels={
        "loan_status": "Loan Status",
        "loan_count": "Number of Loans"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    textposition="outside"
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    legend_title="Loan Status",
    font=dict(size=15),
    xaxis_title="Loan Status",
    yaxis_title="Number of Loans"
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 3
# What is the total loan amount distributed
# across each loan status?
# Soudeep Ghosh
# ==========================================

import pandas as pd
import plotly.express as px

# Load Dataset
loan = pd.read_csv("completedloan.csv")

# ------------------------------------------
# Loan Status Classification
#
# Active:
# C = Running Loan, OK
#
# Closed:
# A = Contract Finished, No Problems
#
# Defaulted:
# B = Contract Finished, Loan Not Paid
# D = Running Loan, Client in Debt
# ------------------------------------------

# Create Loan Status Category
loan["loan_status"] = loan["status"].map({
    "A": "Closed",
    "B": "Defaulted",
    "C": "Active",
    "D": "Defaulted"
})

# ------------------------------------------
# Total Loan Amount by Status
# ------------------------------------------

result = (
    loan.groupby("loan_status")["amount"]
        .sum()
        .reset_index(name="total_loan_amount")
)

# ------------------------------------------
# Sort Result
# ------------------------------------------

status_order = ["Active", "Closed", "Defaulted"]

result["loan_status"] = pd.Categorical(
    result["loan_status"],
    categories=status_order,
    ordered=True
)

result = (
    result.sort_values("loan_status")
          .reset_index(drop=True)
)

# ------------------------------------------
# Display Result
# ------------------------------------------

print("\nQuestion 3 Result")
print(result)

# ------------------------------------------
# Key Findings
# ------------------------------------------

active = result[result["loan_status"] == "Active"].iloc[0]
closed = result[result["loan_status"] == "Closed"].iloc[0]
defaulted = result[result["loan_status"] == "Defaulted"].iloc[0]

print("\nKey Findings")

print(f"Active Loan Amount: {active['total_loan_amount']:,}")
print(f"Closed Loan Amount: {closed['total_loan_amount']:,}")
print(f"Defaulted Loan Amount: {defaulted['total_loan_amount']:,}")

print(f"Total Loan Portfolio: {result['total_loan_amount'].sum():,}")

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.bar(
    result,
    x="loan_status",
    y="total_loan_amount",
    text="total_loan_amount",
    title="Total Loan Amount by Loan Status",
    labels={
        "loan_status": "Loan Status",
        "total_loan_amount": "Total Loan Amount"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    texttemplate="%{text:,}",
    textposition="outside"
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="Loan Status",
    yaxis_title="Total Loan Amount"
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 4
# Has the default rate improved over time?
# Soudeep Ghosh
# ==========================================

import pandas as pd
import plotly.express as px

# Load Dataset
loan = pd.read_csv("completedloan.csv")

# ------------------------------------------
# Loan Default Classification
#
# Defaulted:
# B = Contract Finished, Loan Not Paid
# D = Running Loan, Client in Debt
#
# Non-Defaulted:
# A = Contract Finished, No Problems
# C = Running Loan, OK
# ------------------------------------------

# Create Default Flag
loan["is_default"] = loan["status"].isin(["B", "D"]).astype(int)

# ------------------------------------------
# Total Loans by Year
# ------------------------------------------

total_loans = (
    loan.groupby("year")
        .size()
        .reset_index(name="total_loans")
)

# ------------------------------------------
# Defaulted Loans by Year
# ------------------------------------------

default_loans = (
    loan.groupby("year")["is_default"]
        .sum()
        .reset_index(name="defaulted_loans")
)

# ------------------------------------------
# Merge Results
# ------------------------------------------

result = pd.merge(
    total_loans,
    default_loans,
    on="year"
)

# ------------------------------------------
# Calculate Default Rate
# ------------------------------------------

result["default_rate (%)"] = (
    result["defaulted_loans"] /
    result["total_loans"] * 100
).round(2)

# ------------------------------------------
# Sort by Year
# ------------------------------------------

result = result.sort_values("year").reset_index(drop=True)

# ------------------------------------------
# Display Result
# ------------------------------------------

print("\nQuestion 4 Result")
print(result)

# ------------------------------------------
# Key Findings
# ------------------------------------------

lowest = result.loc[result["default_rate (%)"].idxmin()]
highest = result.loc[result["default_rate (%)"].idxmax()]

print("\nKey Findings")

print(
    f"Highest Default Rate: {highest['default_rate (%)']}% "
    f"in {highest['year']}"
)

print(
    f"Lowest Default Rate: {lowest['default_rate (%)']}% "
    f"in {lowest['year']}"
)

if result["default_rate (%)"].iloc[-1] < result["default_rate (%)"].iloc[0]:
    print("Overall Trend: Default rate has improved over time.")
elif result["default_rate (%)"].iloc[-1] > result["default_rate (%)"].iloc[0]:
    print("Overall Trend: Default rate has worsened over time.")
else:
    print("Overall Trend: Default rate has remained stable over time.")

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.line(
    result,
    x="year",
    y="default_rate (%)",
    markers=True,
    text="default_rate (%)",
    title="Default Rate Over Time",
    labels={
        "year": "Year",
        "default_rate (%)": "Default Rate (%)"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    textposition="top center",
    line=dict(width=3),
    marker=dict(size=10)
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="Year",
    yaxis_title="Default Rate (%)"
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 5
# How has loan repayment progressed over time?
# Soudeep Ghosh
# ==========================================

import pandas as pd
import plotly.express as px

# Load Dataset
loan = pd.read_csv("completedloan.csv")

# ------------------------------------------
# Successfully Repaid Loans
#
# Status A = Contract Finished, No Problems
# ------------------------------------------

# Filter Successfully Repaid Loans
repaid_loans = loan[loan["status"] == "A"]

# ------------------------------------------
# Total Repaid Loan Amount by Year
# ------------------------------------------

result = (
    repaid_loans.groupby("year")["amount"]
    .sum()
    .reset_index(name="repaid_loan_amount")
)

# ------------------------------------------
# Sort by Year
# ------------------------------------------

result = (
    result.sort_values("year")
    .reset_index(drop=True)
)

# ------------------------------------------
# Display Result
# ------------------------------------------

print("\nQuestion 5 Result")
print(result)

# ------------------------------------------
# Key Findings
# ------------------------------------------

highest = result.loc[result["repaid_loan_amount"].idxmax()]
lowest = result.loc[result["repaid_loan_amount"].idxmin()]

print("\nKey Findings")

print(
    f"Highest Repaid Loan Amount: "
    f"{highest['repaid_loan_amount']:,} "
    f"in {int(highest['year'])}"
)

print(
    f"Lowest Repaid Loan Amount: "
    f"{lowest['repaid_loan_amount']:,} "
    f"in {int(lowest['year'])}"
)

print(
    f"Loan repayments peaked in {int(highest['year'])} "
    f"and declined in the following years."
)

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.line(
    result,
    x="year",
    y="repaid_loan_amount",
    markers=True,
    text="repaid_loan_amount",
    title="Loan Repayment Progress Over Time",
    labels={
        "year": "Year",
        "repaid_loan_amount": "Repaid Loan Amount"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    texttemplate="%{text:,}",
    textposition="top center",
    line=dict(width=3),
    marker=dict(size=10)
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="Year",
    yaxis_title="Repaid Loan Amount"
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 7
# Does default rate vary by region?
# SOURABH
# ==========================================

import pandas as pd
import plotly.express as px

# Load Datasets
loan     = pd.read_csv("completedloan.csv")
district = pd.read_csv("completeddistrict.csv")

# Merge Loan with District via location = district_id
merged_df = pd.merge(
    loan,
    district[["district_id", "city", "region"]],
    left_on="location",
    right_on="district_id",
    how="left"
)

# Flag Defaults: D = Default, B = Bad debt
merged_df["is_default"] = merged_df["status"].isin(["D", "B"]).astype(int)

# Default Rate by Region
result = (
    merged_df.groupby("region")
             .agg(
                 total_loans=("loan_id", "count"),
                 defaults=("is_default", "sum")
             )
             .reset_index()
)

result["default_rate_%"] = (
    result["defaults"] / result["total_loans"] * 100
).round(2)

# Sort by Default Rate
result = result.sort_values(
    by="default_rate_%",
    ascending=False
)

print("\nQuestion 7 Result")
print(result.to_string(index=False))

# Key Findings
highest_risk = result.iloc[0]
lowest_risk  = result.iloc[-1]

print("\nKey Findings")

print(
    f"Highest default rate region: "
    f"{highest_risk['region']} "
    f"({highest_risk['default_rate_%']:.2f}% - "
    f"{int(highest_risk['defaults'])} defaults out of {int(highest_risk['total_loans'])} loans)"
)

print(
    f"Lowest default rate region: "
    f"{lowest_risk['region']} "
    f"({lowest_risk['default_rate_%']:.2f}%)"
)

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.bar(
    result,
    x="region",
    y="default_rate_%",
    color="region",
    text="default_rate_%",
    title="Default Rate by Region",
    labels={
        "region": "Region",
        "default_rate_%": "Default Rate (%)"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    texttemplate="%{text:.2f}%",
    textposition="outside"
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="Region",
    yaxis_title="Default Rate (%)",
    showlegend=False
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 8
# Which districts have the highest number of defaulted loans?
# SOURABH
# ==========================================

import pandas as pd
import plotly.express as px

# Load Datasets
loan     = pd.read_csv("completedloan.csv")
district = pd.read_csv("completeddistrict.csv")

# Merge Loan with District
merged_df = pd.merge(
    loan,
    district[["district_id", "city", "region"]],
    left_on="location",
    right_on="district_id",
    how="left"
)

# Filter Defaulted Loans
defaulted = merged_df[
    merged_df["status"].isin(["D", "B"])
]

# Group by District
result = (
    defaulted.groupby(["district_id", "city", "region"])
             .agg(
                 defaulted_loans=("loan_id", "count"),
                 total_defaulted_amount=("amount", "sum")
             )
             .reset_index()
)

# Sort by Defaulted Loans
result = result.sort_values(
    by="defaulted_loans",
    ascending=False
)

print("\nQuestion 8 Result")
print(result.to_string(index=False))

# Key Findings
top_district  = result.iloc[0]
total_defaults = defaulted.shape[0]
top_pct = top_district["defaulted_loans"] / total_defaults * 100

print("\nKey Findings")

print(
    f"Highest default district: "
    f"{top_district['city']} (District {int(top_district['district_id'])}) "
    f"- {int(top_district['defaulted_loans'])} defaults "
    f"({top_pct:.1f}% of all defaults)"
)

print(
    f"Total defaulted amount in top district: "
    f"${int(top_district['total_defaulted_amount']):,}"
)

print(
    f"2nd riskiest district: "
    f"{result.iloc[1]['city']} "
    f"({int(result.iloc[1]['defaulted_loans'])} defaults)"
)

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.bar(
    result,
    x="city",
    y="defaulted_loans",
    color="region",
    text="defaulted_loans",
    title="Defaulted Loans by District",
    labels={
        "city": "District",
        "defaulted_loans": "Number of Defaulted Loans",
        "region": "Region"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    texttemplate="%{text}",
    textposition="outside"
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="District",
    yaxis_title="Number of Defaulted Loans",
    xaxis=dict(tickangle=-45),
    legend_title="Region"
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 9
# Which districts have the highest total loan disbursement?
# SOURABH
# ==========================================

import pandas as pd
import plotly.express as px

# Load Datasets
loan     = pd.read_csv("completedloan.csv")
district = pd.read_csv("completeddistrict.csv")

# Merge Loan with District
merged_df = pd.merge(
    loan,
    district[["district_id", "city", "region"]],
    left_on="location",
    right_on="district_id",
    how="left"
)

# Total Disbursement by District
disbursement = (
    merged_df.groupby(["district_id", "city", "region"])
             .agg(
                 total_loans=("loan_id", "count"),
                 total_disbursed=("amount", "sum"),
                 avg_loan_amount=("amount", "mean")
             )
             .reset_index()
)

disbursement["avg_loan_amount"] = (
    disbursement["avg_loan_amount"].round(2)
)

# Sort by Total Disbursed
disbursement = disbursement.sort_values(
    by="total_disbursed",
    ascending=False
)

# Exposure Share %
disbursement["share_%"] = (
    disbursement["total_disbursed"] /
    disbursement["total_disbursed"].sum() * 100
).round(1)

print("\nQuestion 9 Result")
print(disbursement.to_string(index=False))

# Key Findings
top = disbursement.iloc[0]
total = disbursement["total_disbursed"].sum()
top3_share = disbursement.head(3)["total_disbursed"].sum() / total * 100

print("\nKey Findings")

print(
    f"Highest disbursement district: "
    f"{top['city']} (District {int(top['district_id'])}) "
    f"- ${int(top['total_disbursed']):,} "
    f"({top['share_%']}% of total portfolio)"
)

print(
    f"Top 3 districts combined share: "
    f"{top3_share:.1f}% of total disbursement - extreme exposure concentration"
)

print(
    f"Total portfolio disbursed: ${int(total):,}"
)

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.bar(
    disbursement,
    x="city",
    y="total_disbursed",
    color="region",
    text="total_disbursed",
    title="Total Loan Disbursement by District",
    labels={
        "city": "District",
        "total_disbursed": "Total Loan Disbursed",
        "region": "Region"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    texttemplate="%{text:,}",
    textposition="outside"
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="District",
    yaxis_title="Total Loan Disbursed",
    xaxis=dict(tickangle=-45),
    legend_title="Region"
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 10
# Which loan purposes receive the highest number of loans?
# SOURABH
# ==========================================

import pandas as pd
import plotly.express as px

# Load Dataset
loan = pd.read_csv("completedloan.csv")

# Loan Count by Purpose
loan_count = (
    loan.groupby("purpose")["loan_id"]
        .count()
        .reset_index(name="num_loans")
)

# Total Disbursed by Purpose
total_amount = (
    loan.groupby("purpose")["amount"]
        .sum()
        .reset_index(name="total_amount")
)

# Average Loan Amount by Purpose
avg_amount = (
    loan.groupby("purpose")["amount"]
        .mean()
        .reset_index(name="avg_amount")
)

# Merge Results
result = pd.merge(loan_count, total_amount, on="purpose")
result = pd.merge(result, avg_amount, on="purpose")

result["avg_amount"] = result["avg_amount"].round(2)
result["share_%"] = (
    result["num_loans"] / result["num_loans"].sum() * 100
).round(1)

# Sort by Number of Loans
result = result.sort_values(by="num_loans", ascending=False)

print("\nQuestion 10 Result")
print(result.to_string(index=False))

# Key Findings
top_purpose = result.loc[result["num_loans"].idxmax()]
top_value   = result.loc[result["total_amount"].idxmax()]
highest_avg = result.loc[result["avg_amount"].idxmax()]

print("\nKey Findings")

print(
    f"Most common loan purpose: "
    f"{top_purpose['purpose']} "
    f"({int(top_purpose['num_loans'])} loans - {top_purpose['share_%']}% of total)"
)

print(
    f"Highest total disbursement purpose: "
    f"{top_value['purpose']} "
    f"${int(top_value['total_amount']):,}"
)

print(
    f"Highest avg loan amount purpose: "
    f"{highest_avg['purpose']} "
    f"(${highest_avg['avg_amount']:,.2f} per loan)"
)

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.bar(
    result,
    x="purpose",
    y="num_loans",
    color="purpose",
    text="num_loans",
    title="Number of Loans by Loan Purpose",
    labels={
        "purpose": "Loan Purpose",
        "num_loans": "Number of Loans"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    texttemplate="%{text}",
    textposition="outside"
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="Loan Purpose",
    yaxis_title="Number of Loans",
    showlegend=False
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Performance Analysis
#
# Question 11
# Which loan purposes have the highest default rate?
# Aparupa Chandra
# ===========================================

import pandas as pd
import plotly.express as px

# Load Dataset
loan = pd.read_csv("completedloan.csv")

# Flag which loans count as "defaulted"
# A loan is a default if its status is 'B' or 'D'
loan["is_default"] = loan["status"].isin(["B", "D"])

# Group by purpose and calculate the default rate
# - total_loans   = how many loans exist for that purpose
# - total_defaults = how many of those defaulted
# - default_rate   = total_defaults / total_loans
summary = loan.groupby("purpose")["is_default"].agg(
    total_loans="count",
    total_defaults="sum"
)
summary["default_rate_percent"] = (
        summary["total_defaults"] / summary["total_loans"] * 100
).round(2)

# Sort so the riskiest purpose appears first
summary = summary.sort_values("default_rate_percent", ascending=False)

# Display Result
print("\nQuestion 11 Result")
print("=" * 70)
print("Default rate by loan purpose:")
print("=" * 70)
print(summary)

# Note: 'C' (still running) loans haven't had a chance to default yet,
# so purposes with many recent/active loans may look artificially safer.

# Key Findings
top_purpose = summary.index[0]
top_rate = summary.iloc[0]["default_rate_percent"]
lowest_purpose = summary.index[-1]
lowest_rate = summary.iloc[-1]["default_rate_percent"]
top_loan_count = int(summary.iloc[0]["total_loans"])

print("\n" + "=" * 70)
print("Key Findings")
print("=" * 70)

print(f"- '{top_purpose}' loans have the highest default rate at {top_rate}% "
      f"(based on {top_loan_count} loans), making it the riskiest loan purpose.")
print(f"- '{lowest_purpose}' loans have the lowest default rate at {lowest_rate}%.")
print(f"- The gap between the riskiest and safest purpose is {round(top_rate - lowest_rate, 2)} "
      f"percentage points.")

# ==========================================
# Plotly Visualization
# ==========================================

plot_data = summary.reset_index()

fig = px.bar(
    plot_data,
    x="purpose",
    y="default_rate_percent",
    color="purpose",
    text="default_rate_percent",
    title="Default Rate by Loan Purpose",
    labels={
        "purpose": "Loan Purpose",
        "default_rate_percent": "Default Rate (%)"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    texttemplate="%{text:.2f}%",
    textposition="outside"
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="Loan Purpose",
    yaxis_title="Default Rate (%)",
    showlegend=False
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Performance Analysis
#
# Question 12
# How are loans distributed across different loan durations?
# Aparupa Chandra
# ==========================================

import pandas as pd
import plotly.express as px

# Load Dataset
loan = pd.read_csv("completedloan.csv")

# Count how many loans fall into each duration
duration_counts = loan["duration"].value_counts().sort_index()

# Turning  it into a table with both count and percentage
summary = pd.DataFrame({
    "number_of_loans": duration_counts,
    "percent_of_total": (duration_counts / len(loan) * 100).round(2)
})
summary.index.name = "duration_months"

# Display Result
print("\nQuestion 12 Result")
print("=" * 70)
print(summary)

# Key Findings
most_common = summary["number_of_loans"].idxmax()
most_common_count = summary["number_of_loans"].max()
most_common_pct = summary.loc[most_common, "percent_of_total"]

least_common = summary["number_of_loans"].idxmin()
least_common_count = summary["number_of_loans"].min()

print("\n"+"="*70)
print("Key Findings")
print("="*70)
print(f"- The most common loan duration is {most_common} months, "
      f"with {most_common_count} loans ({most_common_pct}% of all loans).")
print(f"- The least common loan duration is {least_common} months, "
      f"with only {least_common_count} loans.")
print("- Loan counts are fairly evenly spread across all five durations "
      "(12, 24, 36, 48, and 60 months), with no single duration dominating.")

# ==========================================
# Plotly Visualization
# ==========================================

plot_data = summary.reset_index()

fig = px.bar(
    plot_data,
    x="duration_months",
    y="number_of_loans",
    text="number_of_loans",
    title="Loan Distribution by Duration",
    labels={
        "duration_months": "Loan Duration (Months)",
        "number_of_loans": "Number of Loans"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    texttemplate="%{text}",
    textposition="outside"
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="Loan Duration (Months)",
    yaxis_title="Number of Loans"
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Performance Analysis
#
# Question 13
# Does loan length affect default risk?
# Aparupa Chandra
# ==========================================

import pandas as pd
import plotly.express as px

# Load Dataset
loan = pd.read_csv("completedloan.csv")

# Flag which loans count as "defaulted"
# A loan is a default if its status is 'B' or 'D'
loan["is_default"] = loan["status"].isin(["B", "D"])

# Group by duration and calculate the default rate
# - total_loans    = how many loans exist for that duration
# - total_defaults = how many of those defaulted
# - default_rate   = total_defaults / total_loans
summary = loan.groupby("duration")["is_default"].agg(
    total_loans="count",
    total_defaults="sum"
)

summary["default_rate_percent"] = (
    summary["total_defaults"] / summary["total_loans"] * 100
).round(2)

# Keeping durations in natural order (12, 24, 36, 48, 60)
summary = summary.sort_index()

# Display Result
print("\nQuestion 13 Result")
print("=" * 70)
print("Default rate by loan duration (months):")
print("=" * 70)
print(summary)

# Key Findings
shortest = summary.index.min()
shortest_rate = summary.loc[shortest, "default_rate_percent"]
riskiest = summary["default_rate_percent"].idxmax()
riskiest_rate = summary["default_rate_percent"].max()

print("\n" + "=" * 70)
print("Key Findings")
print("=" * 70)

print(f"- The shortest loans ({shortest} months) have the lowest default rate "
      f"at {shortest_rate}%.")

print(f"- The riskiest duration is {riskiest} months, with a default rate of "
      f"{riskiest_rate}%.")

print("- Once loans run longer than 12 months, the default rate jumps and then "
      "stays roughly flat (around 11-12%) regardless of how much longer the "
      "duration gets - so length matters most at the short end, not beyond it.")

# ==========================================
# Plotly Visualization
# ==========================================

plot_data = summary.reset_index()

fig = px.line(
    plot_data,
    x="duration",
    y="default_rate_percent",
    markers=True,
    text="default_rate_percent",
    title="Default Rate by Loan Duration",
    labels={
        "duration": "Loan Duration (Months)",
        "default_rate_percent": "Default Rate (%)"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    texttemplate="%{text:.2f}%",
    textposition="top center",
    line=dict(width=3),
    marker=dict(size=10)
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="Loan Duration (Months)",
    yaxis_title="Default Rate (%)"
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Performance Analysis
#
# Question 14
# What is the relationship between loan amount and monthly repayment?
# Aparupa Chandra
# ==========================================

import pandas as pd
import plotly.express as px

# Load Dataset
loan = pd.read_csv("completedloan.csv")

# Measuring the relationship with a correlation coefficient
correlation = loan["amount"].corr(loan["payments"]).round(3)

# Split loans into 4 equal-sized amount groups (quartiles)
loan["amount_group"] = pd.qcut(loan["amount"], q=4, precision=0)
summary = loan.groupby("amount_group")["payments"].mean().round(1)
summary = summary.rename("average_monthly_payment")

# Display Results
print("\nQuestion 14 Result")
print("=" * 70)
print(f"Correlation between loan amount and monthly payment: {correlation}")
print("=" * 70)
print("\nAverage monthly payment by loan amount group:\n")
print(summary)

# Key Findings
smallest_group_payment = summary.iloc[0]
largest_group_payment = summary.iloc[-1]

print("\n"+"="*70)
print("Key Findings")
print("="*70)
print(f"- Loan amount and monthly payment are strongly positively correlated "
      f"({correlation}) - as the loan gets bigger, the monthly payment goes up too.")
print(f"- The smallest loans average a {smallest_group_payment} monthly payment, "
      f"while the largest loans average {largest_group_payment} - "
      f"nearly {round(largest_group_payment / smallest_group_payment, 1)}x higher.")
print("- The relationship is strong but not perfect (1.0 would be perfect), "
      "because 'duration' also plays a role: a big loan stretched over more "
      "months can still have a manageable monthly payment.")

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.scatter(
    loan,
    x="amount",
    y="payments",
    color="duration",
    title="Relationship Between Loan Amount and Monthly Repayment",
    labels={
        "amount":"Loan Amount",
        "payments":"Monthly Repayment",
        "duration":"Duration (Months)"
    },
    hover_data=["purpose","status"]
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="Loan Amount",
    yaxis_title="Monthly Repayment"
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Performance Analysis
#
# Question 15
# What are the total outstanding loan balance, funded amount,
# and average property value in the luxury loan portfolio?
# Aparupa Chandra
# ===========================================

import pandas as pd
import plotly.express as px

#Load Dataset
luxury = pd.read_csv("luxuryloanportfolio.csv")

# Calculating the three numbers we need
total_outstanding_balance = luxury["loan_balance"].sum()
total_funded_amount = luxury["funded_amount"].sum()
average_property_value = luxury["property_value"].mean()

# Putting the results into one easy-to-read table
summary = pd.DataFrame({
    "metric": [
        "Total Outstanding Loan Balance",
        "Total Funded Amount",
        "Average Property Value"
    ],
    "value": [
        f"${total_outstanding_balance:,.2f}",
        f"${total_funded_amount:,.2f}",
        f"${average_property_value:,.2f}"
    ]
})

# Display Result
print("\nQuestion 15 Result")
print("=" * 70)
print(f"Number of loans in the luxury portfolio: {len(luxury)}\n")
print(summary.to_string(index=False))

# Key Findings
percent_repaid = (1 - total_outstanding_balance / total_funded_amount) * 100

print("\n"+"="*70)
print("Key Findings")
print("="*70)

print(f"- The luxury portfolio has lent out a total of ${total_funded_amount:,.0f}, "
      f"of which ${total_outstanding_balance:,.0f} is still outstanding.")
print(f"- That means roughly {percent_repaid:.1f}% of the funded amount has "
      f"been paid back so far across the portfolio.")
print(f"- The average property backing these loans is worth "
      f"${average_property_value:,.0f}.")

# ==========================================
# Plotly Visualization
# ==========================================

plot_data = pd.DataFrame({
    "metric": [
        "Outstanding Balance",
        "Funded Amount",
        "Average Property Value"
    ],
    "value": [
        total_outstanding_balance,
        total_funded_amount,
        average_property_value
    ]
})

fig = px.bar(
    plot_data,
    x="metric",
    y="value",
    color="metric",
    text="value",
    title="Luxury Loan Portfolio Summary Metrics",
    labels={
        "metric":"Metric",
        "value":"Amount ($)"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    texttemplate="$%{text:,.0f}",
    textposition="outside"
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="Portfolio Metric",
    yaxis_title="Amount ($)",
    showlegend=False
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```
