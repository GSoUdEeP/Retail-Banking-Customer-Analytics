```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 1
# What fraction of loans are healthy vs. problematic?
# Soudeep Ghosh
# ==========================================

import pandas as pd
import plotly.express as px   # Added for Plotly Charts

# Load Dataset
loan = pd.read_csv("completedloan.csv")

# ------------------------------------------
# Create Loan Category
#
# Healthy Loans:
# A = Contract Finished, No Problems
# C = Running Loan, OK
#
# Problematic Loans:
# B = Contract Finished, Loan Not Paid
# D = Running Loan, Client in Debt
# ------------------------------------------

loan["loan_category"] = loan["status"].map({
    "A": "Healthy",
    "B": "Problematic",
    "C": "Healthy",
    "D": "Problematic"
})

# ------------------------------------------
# Count Loans by Category
# ------------------------------------------

result = (
    loan.groupby("loan_category")
        .size()
        .reset_index(name="loan_count")
)

# ------------------------------------------
# Calculate Percentage
# ------------------------------------------

total_loans = result["loan_count"].sum()

result["percentage"] = (
    result["loan_count"] / total_loans * 100
).round(2)

# ------------------------------------------
# Sort Result
# ------------------------------------------

result = result.sort_values(
    by="loan_count",
    ascending=False
).reset_index(drop=True)

# ------------------------------------------
# Display Result
# ------------------------------------------

print("\nQuestion 1 Result")
print(result)

# ------------------------------------------
# Key Findings
# ------------------------------------------

healthy = result[result["loan_category"] == "Healthy"].iloc[0]
problematic = result[result["loan_category"] == "Problematic"].iloc[0]

print("\nKey Findings")

print(
    f"Healthy Loans: {healthy['loan_count']} "
    f"({healthy['percentage']}%)"
)

print(
    f"Problematic Loans: {problematic['loan_count']} "
    f"({problematic['percentage']}%)"
)

print(f"Total Loans: {total_loans}")

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.pie(
    result,
    names="loan_category",          # Slice labels
    values="loan_count",            # Slice values
    title="Healthy vs Problematic Loans",
    hole=0.4                        # Creates a Donut Chart
)

# Display Percentage and Label
fig.update_traces(
    textinfo="percent+label"
)

# Improve Layout
fig.update_layout(
    title_x=0.5,
    legend_title="Loan Category"
)

# Show Chart
fig.show()
```

```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 2
# What is the total number of loans by status
# (Active, Closed, and Defaulted)?
# Soudeep Ghosh
# ==========================================

import pandas as pd
import plotly.express as px

# Load Dataset
loan = pd.read_csv("completedloan.csv")

# ------------------------------------------
# Loan Status Classification
#
# Active:
# C = Running Loan, OK
#
# Closed:
# A = Contract Finished, No Problems
#
# Defaulted:
# B = Contract Finished, Loan Not Paid
# D = Running Loan, Client in Debt
# ------------------------------------------

# Create Loan Status Category
loan["loan_status"] = loan["status"].map({
    "A": "Closed",
    "B": "Defaulted",
    "C": "Active",
    "D": "Defaulted"
})

# ------------------------------------------
# Count Loans by Status
# ------------------------------------------

result = (
    loan.groupby("loan_status")
        .size()
        .reset_index(name="loan_count")
)

# ------------------------------------------
# Sort Result
# ------------------------------------------

status_order = ["Active", "Closed", "Defaulted"]

result["loan_status"] = pd.Categorical(
    result["loan_status"],
    categories=status_order,
    ordered=True
)

result = (
    result.sort_values("loan_status")
          .reset_index(drop=True)
)

# ------------------------------------------
# Display Result
# ------------------------------------------

print("\nQuestion 2 Result")
print(result)

# ------------------------------------------
# Key Findings
# ------------------------------------------

active = result[result["loan_status"] == "Active"].iloc[0]
closed = result[result["loan_status"] == "Closed"].iloc[0]
defaulted = result[result["loan_status"] == "Defaulted"].iloc[0]

print("\nKey Findings")

print(f"Active Loans: {active['loan_count']}")
print(f"Closed Loans: {closed['loan_count']}")
print(f"Defaulted Loans: {defaulted['loan_count']}")
print(f"Total Loans: {result['loan_count'].sum()}")

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.bar(
    result,
    x="loan_status",
    y="loan_count",
    text="loan_count",
    title="Total Number of Loans by Status",
    labels={
        "loan_status": "Loan Status",
        "loan_count": "Number of Loans"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    textposition="outside"
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    legend_title="Loan Status",
    font=dict(size=15),
    xaxis_title="Loan Status",
    yaxis_title="Number of Loans"
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 3
# What is the total loan amount distributed
# across each loan status?
# Soudeep Ghosh
# ==========================================

import pandas as pd
import plotly.express as px

# Load Dataset
loan = pd.read_csv("completedloan.csv")

# ------------------------------------------
# Loan Status Classification
#
# Active:
# C = Running Loan, OK
#
# Closed:
# A = Contract Finished, No Problems
#
# Defaulted:
# B = Contract Finished, Loan Not Paid
# D = Running Loan, Client in Debt
# ------------------------------------------

# Create Loan Status Category
loan["loan_status"] = loan["status"].map({
    "A": "Closed",
    "B": "Defaulted",
    "C": "Active",
    "D": "Defaulted"
})

# ------------------------------------------
# Total Loan Amount by Status
# ------------------------------------------

result = (
    loan.groupby("loan_status")["amount"]
        .sum()
        .reset_index(name="total_loan_amount")
)

# ------------------------------------------
# Sort Result
# ------------------------------------------

status_order = ["Active", "Closed", "Defaulted"]

result["loan_status"] = pd.Categorical(
    result["loan_status"],
    categories=status_order,
    ordered=True
)

result = (
    result.sort_values("loan_status")
          .reset_index(drop=True)
)

# ------------------------------------------
# Display Result
# ------------------------------------------

print("\nQuestion 3 Result")
print(result)

# ------------------------------------------
# Key Findings
# ------------------------------------------

active = result[result["loan_status"] == "Active"].iloc[0]
closed = result[result["loan_status"] == "Closed"].iloc[0]
defaulted = result[result["loan_status"] == "Defaulted"].iloc[0]

print("\nKey Findings")

print(f"Active Loan Amount: {active['total_loan_amount']:,}")
print(f"Closed Loan Amount: {closed['total_loan_amount']:,}")
print(f"Defaulted Loan Amount: {defaulted['total_loan_amount']:,}")

print(f"Total Loan Portfolio: {result['total_loan_amount'].sum():,}")

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.bar(
    result,
    x="loan_status",
    y="total_loan_amount",
    text="total_loan_amount",
    title="Total Loan Amount by Loan Status",
    labels={
        "loan_status": "Loan Status",
        "total_loan_amount": "Total Loan Amount"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    texttemplate="%{text:,}",
    textposition="outside"
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="Loan Status",
    yaxis_title="Total Loan Amount"
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 4
# Has the default rate improved over time?
# Soudeep Ghosh
# ==========================================

import pandas as pd
import plotly.express as px

# Load Dataset
loan = pd.read_csv("completedloan.csv")

# ------------------------------------------
# Loan Default Classification
#
# Defaulted:
# B = Contract Finished, Loan Not Paid
# D = Running Loan, Client in Debt
#
# Non-Defaulted:
# A = Contract Finished, No Problems
# C = Running Loan, OK
# ------------------------------------------

# Create Default Flag
loan["is_default"] = loan["status"].isin(["B", "D"]).astype(int)

# ------------------------------------------
# Total Loans by Year
# ------------------------------------------

total_loans = (
    loan.groupby("year")
        .size()
        .reset_index(name="total_loans")
)

# ------------------------------------------
# Defaulted Loans by Year
# ------------------------------------------

default_loans = (
    loan.groupby("year")["is_default"]
        .sum()
        .reset_index(name="defaulted_loans")
)

# ------------------------------------------
# Merge Results
# ------------------------------------------

result = pd.merge(
    total_loans,
    default_loans,
    on="year"
)

# ------------------------------------------
# Calculate Default Rate
# ------------------------------------------

result["default_rate (%)"] = (
    result["defaulted_loans"] /
    result["total_loans"] * 100
).round(2)

# ------------------------------------------
# Sort by Year
# ------------------------------------------

result = result.sort_values("year").reset_index(drop=True)

# ------------------------------------------
# Display Result
# ------------------------------------------

print("\nQuestion 4 Result")
print(result)

# ------------------------------------------
# Key Findings
# ------------------------------------------

lowest = result.loc[result["default_rate (%)"].idxmin()]
highest = result.loc[result["default_rate (%)"].idxmax()]

print("\nKey Findings")

print(
    f"Highest Default Rate: {highest['default_rate (%)']}% "
    f"in {highest['year']}"
)

print(
    f"Lowest Default Rate: {lowest['default_rate (%)']}% "
    f"in {lowest['year']}"
)

if result["default_rate (%)"].iloc[-1] < result["default_rate (%)"].iloc[0]:
    print("Overall Trend: Default rate has improved over time.")
elif result["default_rate (%)"].iloc[-1] > result["default_rate (%)"].iloc[0]:
    print("Overall Trend: Default rate has worsened over time.")
else:
    print("Overall Trend: Default rate has remained stable over time.")

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.line(
    result,
    x="year",
    y="default_rate (%)",
    markers=True,
    text="default_rate (%)",
    title="Default Rate Over Time",
    labels={
        "year": "Year",
        "default_rate (%)": "Default Rate (%)"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    textposition="top center",
    line=dict(width=3),
    marker=dict(size=10)
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="Year",
    yaxis_title="Default Rate (%)"
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 5
# How has loan repayment progressed over time?
# Soudeep Ghosh
# ==========================================

import pandas as pd
import plotly.express as px

# Load Dataset
loan = pd.read_csv("completedloan.csv")

# ------------------------------------------
# Successfully Repaid Loans
#
# Status A = Contract Finished, No Problems
# ------------------------------------------

# Filter Successfully Repaid Loans
repaid_loans = loan[loan["status"] == "A"]

# ------------------------------------------
# Total Repaid Loan Amount by Year
# ------------------------------------------

result = (
    repaid_loans.groupby("year")["amount"]
    .sum()
    .reset_index(name="repaid_loan_amount")
)

# ------------------------------------------
# Sort by Year
# ------------------------------------------

result = (
    result.sort_values("year")
    .reset_index(drop=True)
)

# ------------------------------------------
# Display Result
# ------------------------------------------

print("\nQuestion 5 Result")
print(result)

# ------------------------------------------
# Key Findings
# ------------------------------------------

highest = result.loc[result["repaid_loan_amount"].idxmax()]
lowest = result.loc[result["repaid_loan_amount"].idxmin()]

print("\nKey Findings")

print(
    f"Highest Repaid Loan Amount: "
    f"{highest['repaid_loan_amount']:,} "
    f"in {int(highest['year'])}"
)

print(
    f"Lowest Repaid Loan Amount: "
    f"{lowest['repaid_loan_amount']:,} "
    f"in {int(lowest['year'])}"
)

print(
    f"Loan repayments peaked in {int(highest['year'])} "
    f"and declined in the following years."
)

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.line(
    result,
    x="year",
    y="repaid_loan_amount",
    markers=True,
    text="repaid_loan_amount",
    title="Loan Repayment Progress Over Time",
    labels={
        "year": "Year",
        "repaid_loan_amount": "Repaid Loan Amount"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    texttemplate="%{text:,}",
    textposition="top center",
    line=dict(width=3),
    marker=dict(size=10)
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="Year",
    yaxis_title="Repaid Loan Amount"
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 7
# Does default rate vary by region?
# SOURABH
# ==========================================

import pandas as pd
import plotly.express as px

# Load Datasets
loan     = pd.read_csv("completedloan.csv")
district = pd.read_csv("completeddistrict.csv")

# Merge Loan with District via location = district_id
merged_df = pd.merge(
    loan,
    district[["district_id", "city", "region"]],
    left_on="location",
    right_on="district_id",
    how="left"
)

# Flag Defaults: D = Default, B = Bad debt
merged_df["is_default"] = merged_df["status"].isin(["D", "B"]).astype(int)

# Default Rate by Region
result = (
    merged_df.groupby("region")
             .agg(
                 total_loans=("loan_id", "count"),
                 defaults=("is_default", "sum")
             )
             .reset_index()
)

result["default_rate_%"] = (
    result["defaults"] / result["total_loans"] * 100
).round(2)

# Sort by Default Rate
result = result.sort_values(
    by="default_rate_%",
    ascending=False
)

print("\nQuestion 7 Result")
print(result.to_string(index=False))

# Key Findings
highest_risk = result.iloc[0]
lowest_risk  = result.iloc[-1]

print("\nKey Findings")

print(
    f"Highest default rate region: "
    f"{highest_risk['region']} "
    f"({highest_risk['default_rate_%']:.2f}% - "
    f"{int(highest_risk['defaults'])} defaults out of {int(highest_risk['total_loans'])} loans)"
)

print(
    f"Lowest default rate region: "
    f"{lowest_risk['region']} "
    f"({lowest_risk['default_rate_%']:.2f}%)"
)

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.bar(
    result,
    x="region",
    y="default_rate_%",
    color="region",
    text="default_rate_%",
    title="Default Rate by Region",
    labels={
        "region": "Region",
        "default_rate_%": "Default Rate (%)"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    texttemplate="%{text:.2f}%",
    textposition="outside"
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="Region",
    yaxis_title="Default Rate (%)",
    showlegend=False
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 8
# Which districts have the highest number of defaulted loans?
# SOURABH
# ==========================================

import pandas as pd
import plotly.express as px

# Load Datasets
loan     = pd.read_csv("completedloan.csv")
district = pd.read_csv("completeddistrict.csv")

# Merge Loan with District
merged_df = pd.merge(
    loan,
    district[["district_id", "city", "region"]],
    left_on="location",
    right_on="district_id",
    how="left"
)

# Filter Defaulted Loans
defaulted = merged_df[
    merged_df["status"].isin(["D", "B"])
]

# Group by District
result = (
    defaulted.groupby(["district_id", "city", "region"])
             .agg(
                 defaulted_loans=("loan_id", "count"),
                 total_defaulted_amount=("amount", "sum")
             )
             .reset_index()
)

# Sort by Defaulted Loans
result = result.sort_values(
    by="defaulted_loans",
    ascending=False
)

print("\nQuestion 8 Result")
print(result.to_string(index=False))

# Key Findings
top_district  = result.iloc[0]
total_defaults = defaulted.shape[0]
top_pct = top_district["defaulted_loans"] / total_defaults * 100

print("\nKey Findings")

print(
    f"Highest default district: "
    f"{top_district['city']} (District {int(top_district['district_id'])}) "
    f"- {int(top_district['defaulted_loans'])} defaults "
    f"({top_pct:.1f}% of all defaults)"
)

print(
    f"Total defaulted amount in top district: "
    f"${int(top_district['total_defaulted_amount']):,}"
)

print(
    f"2nd riskiest district: "
    f"{result.iloc[1]['city']} "
    f"({int(result.iloc[1]['defaulted_loans'])} defaults)"
)

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.bar(
    result,
    x="city",
    y="defaulted_loans",
    color="region",
    text="defaulted_loans",
    title="Defaulted Loans by District",
    labels={
        "city": "District",
        "defaulted_loans": "Number of Defaulted Loans",
        "region": "Region"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    texttemplate="%{text}",
    textposition="outside"
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="District",
    yaxis_title="Number of Defaulted Loans",
    xaxis=dict(tickangle=-45),
    legend_title="Region"
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Portfolio Analysis
#
# Question 9
# Which districts have the highest total loan disbursement?
# SOURABH
# ==========================================

import pandas as pd
import plotly.express as px

# Load Datasets
loan     = pd.read_csv("completedloan.csv")
district = pd.read_csv("completeddistrict.csv")

# Merge Loan with District
merged_df = pd.merge(
    loan,
    district[["district_id", "city", "region"]],
    left_on="location",
    right_on="district_id",
    how="left"
)

# Total Disbursement by District
disbursement = (
    merged_df.groupby(["district_id", "city", "region"])
             .agg(
                 total_loans=("loan_id", "count"),
                 total_disbursed=("amount", "sum"),
                 avg_loan_amount=("amount", "mean")
             )
             .reset_index()
)

disbursement["avg_loan_amount"] = (
    disbursement["avg_loan_amount"].round(2)
)

# Sort by Total Disbursed
disbursement = disbursement.sort_values(
    by="total_disbursed",
    ascending=False
)

# Exposure Share %
disbursement["share_%"] = (
    disbursement["total_disbursed"] /
    disbursement["total_disbursed"].sum() * 100
).round(1)

print("\nQuestion 9 Result")
print(disbursement.to_string(index=False))

# Key Findings
top = disbursement.iloc[0]
total = disbursement["total_disbursed"].sum()
top3_share = disbursement.head(3)["total_disbursed"].sum() / total * 100

print("\nKey Findings")

print(
    f"Highest disbursement district: "
    f"{top['city']} (District {int(top['district_id'])}) "
    f"- ${int(top['total_disbursed']):,} "
    f"({top['share_%']}% of total portfolio)"
)

print(
    f"Top 3 districts combined share: "
    f"{top3_share:.1f}% of total disbursement - extreme exposure concentration"
)

print(
    f"Total portfolio disbursed: ${int(total):,}"
)

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.bar(
    disbursement,
    x="city",
    y="total_disbursed",
    color="region",
    text="total_disbursed",
    title="Total Loan Disbursement by District",
    labels={
        "city": "District",
        "total_disbursed": "Total Loan Disbursed",
        "region": "Region"
    }
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_traces(
    texttemplate="%{text:,}",
    textposition="outside"
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="District",
    yaxis_title="Total Loan Disbursed",
    xaxis=dict(tickangle=-45),
    legend_title="Region"
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 03
# Loan Performance Analysis
#
# Question 14
# What is the relationship between loan amount and monthly repayment?
# Aparupa Chandra
# ==========================================

import pandas as pd
import plotly.express as px

# Load Dataset
loan = pd.read_csv("completedloan.csv")

# Measuring the relationship with a correlation coefficient
correlation = loan["amount"].corr(loan["payments"]).round(3)

# Split loans into 4 equal-sized amount groups (quartiles)
loan["amount_group"] = pd.qcut(loan["amount"], q=4, precision=0)
summary = loan.groupby("amount_group")["payments"].mean().round(1)
summary = summary.rename("average_monthly_payment")

# Display Results
print("\nQuestion 14 Result")
print("=" * 70)
print(f"Correlation between loan amount and monthly payment: {correlation}")
print("=" * 70)
print("\nAverage monthly payment by loan amount group:\n")
print(summary)

# Key Findings
smallest_group_payment = summary.iloc[0]
largest_group_payment = summary.iloc[-1]

print("\n"+"="*70)
print("Key Findings")
print("="*70)
print(f"- Loan amount and monthly payment are strongly positively correlated "
      f"({correlation}) - as the loan gets bigger, the monthly payment goes up too.")
print(f"- The smallest loans average a {smallest_group_payment} monthly payment, "
      f"while the largest loans average {largest_group_payment} - "
      f"nearly {round(largest_group_payment / smallest_group_payment, 1)}x higher.")
print("- The relationship is strong but not perfect (1.0 would be perfect), "
      "because 'duration' also plays a role: a big loan stretched over more "
      "months can still have a manageable monthly payment.")

# ==========================================
# Plotly Visualization
# ==========================================

fig = px.scatter(
    loan,
    x="amount",
    y="payments",
    color="duration",
    title="Relationship Between Loan Amount and Monthly Repayment",
    labels={
        "amount":"Loan Amount",
        "payments":"Monthly Repayment",
        "duration":"Duration (Months)"
    },
    hover_data=["purpose","status"]
)

# ------------------------------------------
# Professional Formatting
# ------------------------------------------

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="Loan Amount",
    yaxis_title="Monthly Repayment"
)

# ------------------------------------------
# Display Chart
# ------------------------------------------

fig.show()
```

```python
# ==========================================
# Page 04
# CRM & Customer Experience
#
# Theme 2: Performance & SLA
# Question 4
# Do high-priority complaints get resolved faster, and is resolution rate actually higher for high priority?
# Sourabh Bhattacharya
# ==========================================



import pandas as pd
import plotly.express as px

# Load Dataset
df = pd.read_csv("crm_call_center_logs.csv")

# Keep only interpretable priority levels
df = df[df["priority"].isin([0, 1, 2])].copy()

# Label Priorities
df["priority_label"] = df["priority"].map({
    0: "Low (0)",
    1: "Medium (1)",
    2: "High (2)"
})

# Average Resolution Time by Priority
resolution_time = (
    df.groupby("priority_label")["service_seconds"]
      .mean()
      .reset_index(name="avg_service_seconds")
)

# Resolution Rate (Agent outcome = resolved)
df["is_resolved"] = (
    df["outcome"] == "Agent"
).astype(int)

resolution_rate = (
    df.groupby("priority_label")["is_resolved"]
      .mean()
      .reset_index(name="resolution_rate_pct")
)

resolution_rate["resolution_rate_pct"] = (
    resolution_rate["resolution_rate_pct"] * 100
).round(2)

# Merge Results
result = pd.merge(
    resolution_time,
    resolution_rate,
    on="priority_label"
)

# Round Values
result["avg_service_seconds"] = (
    result["avg_service_seconds"]
    .round(2)
)

# Sort by Priority
priority_order = ["Low (0)", "Medium (1)", "High (2)"]
result["priority_label"] = pd.Categorical(
    result["priority_label"],
    categories=priority_order,
    ordered=True
)
result = result.sort_values("priority_label")

# Display Result
print("\nQuestion 4 Result")
print(result)

# Key Findings
fastest = result.loc[
    result["avg_service_seconds"].idxmin()
]

highest_res = result.loc[
    result["resolution_rate_pct"].idxmax()
]

print("\nKey Findings")

print(
    f"Priority level with fastest resolution: "
    f"{fastest['priority_label']} "
    f"({fastest['avg_service_seconds']:.2f} seconds)"
)

print(
    f"Priority level with highest resolution rate: "
    f"{highest_res['priority_label']} "
    f"({highest_res['resolution_rate_pct']:.2f}%)"
)

# ==========================================
# Plotly Visualization
# ==========================================

# Create Grouped Bar Chart
fig = px.bar(
    result,
    x="priority_label",
    y=["avg_service_seconds", "resolution_rate_pct"],
    barmode="group",
    text_auto=".2s",
    title="Resolution Time and Resolution Rate by Priority Level",
    labels={
        "priority_label": "Priority Level",
        "value": "Seconds / Percentage (%)",
        "variable": "Metric"
    }
)

# Professional Dashboard Formatting
fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    font=dict(size=15),
    xaxis_title="Priority Level",
    yaxis_title="Avg Resolution Time (Seconds) / Resolution Rate (%)",
    legend_title="Metrics"
)

# Display Values Outside Bars
fig.update_traces(
    textposition="outside"
)

# Show Chart
fig.show()
```

```python
# ==========================================
# Page 04
# CRM & Customer Experience
#
# Theme 4: Product & Satisfaction
# Question 11
# Which products perform worst overall, combining complaint count, resolution rate, timely response, and rating?
# Aparupa Chandra
# ==========================================
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

#Load Dataset
EVENTS_PATH  = "crm_events.csv"
REVIEWS_PATH = "crm_reviews.csv"

# Manual Mapping: Review-Brand -> Complaint-Category
PRODUCT_MAP = {
    "Eagle National Bank":     "Bank account or service",
    "Eagle National Mortgage": "Credit card",   # ASSUMPTION - verify
    "Eagle Capital":           "Credit card",   # ASSUMPTION - verify
}

# COMPLAINT-SIDE METRICS (from crm_events.csv)
events = pd.read_csv(EVENTS_PATH)

# Resolution flag: counts as "resolved" only if it actually closed with
# some form of relief for the customer
RESOLVED_STATUSES = [
    "Closed with relief",
    "Closed with monetary relief",
    "Closed with non-monetary relief",
]
events["is_resolved"] = events["company_response_to_consumer"].isin(RESOLVED_STATUSES)
events["is_timely"] = events["timely_response"].str.strip().str.lower() == "yes"

complaint_metrics = (
    events.groupby("product")
    .agg(
        complaint_count=("complaint_id", "count"),
        resolution_rate=("is_resolved", "mean"),
        timely_response_rate=("is_timely", "mean"),
    )
    .reset_index()
)
complaint_metrics["resolution_rate"] = (complaint_metrics["resolution_rate"] * 100).round(2)
complaint_metrics["timely_response_rate"] = (complaint_metrics["timely_response_rate"] * 100).round(2)

#Display Result
print("Question 11 Result")
print("=" * 70)
print("COMPLAINT-SIDE METRICS BY PRODUCT (crm_events.csv)")
print("=" * 70)
print(complaint_metrics.to_string(index=False))

# REVIEW-SIDE METRICS (from crm_reviews.csv)
reviews = pd.read_csv(REVIEWS_PATH)

review_metrics = (
    reviews.groupby("product")
    .agg(
        review_count=("stars", "count"),
        avg_rating=("stars", "mean"),
    )
    .reset_index()
)
review_metrics["avg_rating"] = review_metrics["avg_rating"].round(2)
print("\n" + "=" * 70)
print("REVIEW-SIDE METRICS BY PRODUCT (crm_reviews.csv)")
print("=" * 70)
print(review_metrics.to_string(index=False))

# COMBINED "WORST PERFORMER" SCORE — relies on PRODUCT_MAP assumption
review_metrics["mapped_product"] = review_metrics["product"].map(PRODUCT_MAP)

# Aggregate review metrics up to the complaint-category level
review_rollup = (
    review_metrics.groupby("mapped_product")
    .agg(
        review_count=("review_count", "sum"),
        avg_rating=("avg_rating", "mean"),  # simple mean of brand averages
    )
    .reset_index()
    .rename(columns={"mapped_product": "product"})
)

combined = complaint_metrics.merge(review_rollup, on="product", how="left")

# Build a composite "worst performer" score:
# higher complaint_count = worse | lower resolution_rate = worse
# lower timely_response_rate = worse | lower avg_rating = worse
def normalize(series, invert=False):
    norm = (series - series.min()) / (series.max() - series.min())
    return 1 - norm if invert else norm

combined["score_complaint_volume"] = normalize(combined["complaint_count"])          # high = bad
combined["score_resolution"] = normalize(combined["resolution_rate"], invert=True)    # low rate = bad
combined["score_timeliness"] = normalize(combined["timely_response_rate"], invert=True)
combined["score_rating"] = normalize(combined["avg_rating"], invert=True)

combined["worst_performer_score"] = combined[
    ["score_complaint_volume", "score_resolution", "score_timeliness", "score_rating"]
].mean(axis=1).round(3)

combined = combined.sort_values("worst_performer_score", ascending=False)

#Key Findings
print("\nKey Findings")
print("=" * 70)
print("COMBINED VIEW-based on PRODUCT_MAP assumption")
print("Higher 'worst_performer_score'= worse overall performance")
print("=" * 70)
print(
    combined[
        ["product", "complaint_count", "resolution_rate", "timely_response_rate",
         "avg_rating", "worst_performer_score"]
    ].to_string(index=False)
)


# ============================================================
# Plotly Visualization
# ============================================================
print("\n" + "=" * 70)
print("RENDERING VISUALIZATIONS")
print("=" * 70)

# Display Chart 1: Complaint volume (bars) vs resolution/timeliness rate (lines)

fig1 = go.Figure()
fig1.add_bar(
    x=complaint_metrics["product"],
    y=complaint_metrics["complaint_count"],
    name="Complaint Count",
    marker_color="#1f77b4",
    yaxis="y1",
)
fig1.add_trace(go.Scatter(
    x=complaint_metrics["product"],
    y=complaint_metrics["resolution_rate"],
    name="Resolution Rate (%)",
    mode="lines+markers",
    yaxis="y2",
    line=dict(color="#2ca02c", width=3),
))
fig1.add_trace(go.Scatter(
    x=complaint_metrics["product"],
    y=complaint_metrics["timely_response_rate"],
    name="Timely Response Rate (%)",
    mode="lines+markers",
    yaxis="y2",
    line=dict(color="#ff7f0e", width=3, dash="dot"),
))
fig1.update_layout(
    xaxis_title="Product",
    yaxis=dict(title="Complaint Count"),
    yaxis2=dict(title="Rate (%)", overlaying="y", side="right", range=[0, 100]),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
    template="plotly_white",
    margin=dict(t=60),
)
fig1.show()

# Display Chart 2: Average rating by product (label = review count)

fig2 = px.bar(
    review_metrics,
    x="product",
    y="avg_rating",
    text="review_count",
    color="avg_rating",
    color_continuous_scale="RdYlGn",
    labels={"avg_rating": "Average Rating (stars)", "product": "Product"},
)
fig2.update_traces(texttemplate="n=%{text}", textposition="outside")
fig2.update_layout(
    template="plotly_white",
    coloraxis_showscale=False,
    yaxis_range=[0, 5],
)
fig2.show()

# Display Chart 3: Composite "worst performer" ranking
# Composite ranking below is based on the PRODUCT_MAP assumption
fig3 = px.bar(
    combined,
    x="worst_performer_score",
    y="product",
    orientation="h",
    color="worst_performer_score",
    color_continuous_scale="Reds",
    text="worst_performer_score",
    labels={"worst_performer_score": "Worst Performer Score", "product": "Product"},
)
fig3.update_traces(texttemplate="%{text:.3f}", textposition="outside")
fig3.update_layout(
    template="plotly_white",
    yaxis=dict(autorange="reversed"),  # keeping  worst-first order from the dataframe
    coloraxis_showscale=False,
    margin=dict(t=20),
)
fig3.show()
```

```python
# ==========================================
# Page 04
# CRM & Customer Experience
#
# Theme 4: Product & Satisfaction
# Question 12
# What's the star rating distribution, and is there a link between complaint volume and rating for a product?
# Aparupa Chandra
# ==========================================
import pandas as pd
import plotly.express as px

# Load Dataset
REVIEWS_PATH = "crm_reviews.csv"
EVENTS_PATH  = "crm_events.csv"

# Manual Mapping: Review-Brand -> Complaint-Category
PRODUCT_MAP = {
"Eagle National Bank":     "Bank account or service",
"Eagle National Mortgage": "Credit card",   # ASSUMPTION - verify
"Eagle Capital":           "Credit card",   # ASSUMPTION - verify
}

# STAR RATING DISTRIBUTION — overall and by product
reviews = pd.read_csv(REVIEWS_PATH)

# Display Result
print("\n"+"\n"+"=" * 70)
print("Question  12 Result")
print("=" * 70)
print("OVERALL STAR RATING DISTRIBUTION")
print("=" * 70)
overall_dist = reviews["stars"].value_counts().sort_index()
overall_pct = (overall_dist / overall_dist.sum() * 100).round(2)
dist_table = pd.DataFrame({"count": overall_dist, "pct_of_total": overall_pct})
print(dist_table.to_string())

print("\n" + "=" * 70)
print("STAR RATING DISTRIBUTION BY PRODUCT (brand)")
print("=" * 70)
by_product_dist = (
reviews.groupby(["product", "stars"]).size().unstack(fill_value=0)
)
print(by_product_dist.to_string())

print("\n" + "=" * 70)
print("AVG RATING & REVIEW COUNT BY PRODUCT (brand)")
print("=" * 70)
brand_summary = (
reviews.groupby("product")
.agg(review_count=("stars", "count"), avg_rating=("stars", "mean"))
.reset_index()
)
brand_summary["avg_rating"] = brand_summary["avg_rating"].round(2)
print(brand_summary.to_string(index=False))

# COMPLAINT VOLUME BY PRODUCT (from crm_events.csv)
events = pd.read_csv(EVENTS_PATH)

complaint_volume = (
events.groupby("product").size().reset_index(name="complaint_count")
)

print("\n" + "=" * 70)
print("COMPLAINT VOLUME BY PRODUCT (complaint category)")
print("=" * 70)
print(complaint_volume.to_string(index=False))

# LINK BETWEEN COMPLAINT VOLUME AND RATING — via PRODUCT_MAP

brand_summary["mapped_category"] = brand_summary["product"].map(PRODUCT_MAP)
merged = brand_summary.merge(
complaint_volume, left_on="mapped_category", right_on="product", suffixes=("_brand", "_category")
)

print("\n" + "=" * 70)
print("COMPLAINT VOLUME vs AVG RATING (mapped, per brand)")
print("CAUTION: only", len(merged), "data points-correlation is illustrative only")
print("=" * 70)
view = merged[["product_brand", "mapped_category", "review_count", "avg_rating", "complaint_count"]]
print(view.to_string(index=False))

# Key Findinds
if len(merged) >= 2:
    corr = merged["avg_rating"].corr(merged["complaint_count"])
    print("\nKey Findings")
    print("=" * 70)
    print(f"Pearson correlation (avg_rating vs complaint_count): {corr:.3f}")
    if corr < -0.3:
        note = "Negative link: higher complaint volume tends to come with lower ratings."
    elif corr > 0.3:
        note = "Positive link: higher complaint volume tends to come with HIGHER ratings (counterintuitive — may just mean bigger/more visible products get both more complaints and more reviews)."
    else:
        note = "No strong linear link detected between complaint volume and rating."
    print(note)
    print("NOTE: with only", len(merged), "points this is NOT statistically reliable — treating as a starting hypothesis, not a conclusion.")
else:
    print("\nNot enough mapped data points to compute a correlation.")


# ============================================================
# Plotly Visualizations
# ============================================================
print("\n" + "=" * 70)
print("RENDERING VISUALIZATIONS")
print("=" * 70)

# Display Chart 1: Overall star rating distribution
dist_plot_df = dist_table.reset_index()  # columns: stars, count, pct_of_total

fig1 = px.bar(
    dist_plot_df,
    x="stars",
    y="count",
    text="pct_of_total",
    color="stars",
    color_continuous_scale="RdYlGn",
    title="Overall Star Rating Distribution",
    labels={"stars": "Star Rating", "count": "Number of Reviews"},
)
fig1.update_traces(texttemplate="%{text}%", textposition="outside")
fig1.update_layout(
    template="plotly_white",
    xaxis=dict(dtick=1),
    coloraxis_showscale=False,
)
fig1.show()

# Display Chart 2: Star rating distribution by product (100% stacked bar)

by_product_pct = by_product_dist.div(by_product_dist.sum(axis=1), axis=0) * 100
by_product_pct_long = (
    by_product_pct.reset_index()
    .melt(id_vars="product", var_name="stars", value_name="pct")
)
by_product_pct_long["stars"] = by_product_pct_long["stars"].astype(str)

star_colors = {
    "1": "#d73027",
    "2": "#fc8d59",
    "3": "#fee08b",
    "4": "#91cf60",
    "5": "#1a9850",
}

fig2 = px.bar(
    by_product_pct_long,
    x="product",
    y="pct",
    color="stars",
    color_discrete_map=star_colors,
    category_orders={"stars": ["1", "2", "3", "4", "5"]},
    title="Star Rating Distribution by Product (% of reviews)",
    labels={"pct": "% of Reviews", "product": "Product", "stars": "Star Rating"},
)
fig2.update_layout(
    template="plotly_white",
    barmode="stack",
    yaxis_range=[0, 100],
)
fig2.show()

# Display Chart 3: Complaint volume vs avg rating (the "link" question)

fig3 = px.scatter(
    merged,
    x="complaint_count",
    y="avg_rating",
    size="review_count",
    color="product_brand",
    text="product_brand",
    title="Complaint Volume vs Average Rating by Product (bubble size = review count)",
    labels={
        "complaint_count": "Complaint Count",
        "avg_rating": "Average Rating (stars)",
        "product_brand": "Product",
    },
)
# Note: add trendline="ols" to px.scatter above if you have the
# 'statsmodels' package installed and want a fitted trend line.
fig3.update_traces(textposition="top center")
fig3.update_layout(
    template="plotly_white",
    yaxis_range=[0, 5],
    showlegend=False,
)
fig3.show()
```

```python
# ==========================================
# Page 04
# CRM & Customer Experience
#
# Theme 5: Geography,Trends & Behavior
# Question 13
# Which districts have the highest complaint volume and lowest satisfaction — are there clear outliers?
# Aparupa Chandra
# ==========================================

import pandas as pd
import plotly.express as px

# Load Dataset
df = pd.read_csv("crm_reviews.csv")
# DISTRICT-LEVEL SUMMARY: review volume + avg rating
district_summary = (df.groupby("district_id").agg(
            review_volume=("stars", "count"),
            avg_rating=("stars", "mean"),
            pct_1_star=("stars", lambda s: (s == 1).mean() * 100),
        )
        .reset_index()
    )
district_summary["avg_rating"] = district_summary["avg_rating"].round(2)
district_summary["pct_1_star"] = district_summary["pct_1_star"].round(1)

#Display Result
print("\n"+"\n"+"=" * 70)
print("Question  13 Result")
print("=" * 70)
print(f"Total districts: {district_summary.shape[0]}")
print(f"Review volume per district-min: {district_summary['review_volume'].min()}, "
          f"max: {district_summary['review_volume'].max()}, "
          f"median: {district_summary['review_volume'].median()}")
print("=" * 70)

# TOP DISTRICTS BY VOLUME (highest review/complaint-proxy volume)
print("\n" + "=" * 70)
print("TOP 10 DISTRICTS BY REVIEW VOLUME")
print("=" * 70)
print(
    district_summary.sort_values("review_volume", ascending=False)
    .head(10)
    .to_string(index=False)
)

# BOTTOM DISTRICTS BY SATISFACTION (lowest avg rating)
print("\n" + "=" * 70)
print("BOTTOM 10 DISTRICTS BY AVG RATING (lowest satisfaction)")
print("=" * 70)
print(
    district_summary.sort_values("avg_rating", ascending=True)
    .head(10)
    .to_string(index=False)
)

# OUTLIER DETECTION — districts that are BOTH high-volume AND
#    low-satisfaction (the genuinely concerning ones)
vol_threshold = district_summary["review_volume"].quantile(0.75)  # top quartile = "high volume"
rating_threshold = district_summary["avg_rating"].quantile(0.25)  # bottom quartile = "low satisfaction"

print("\n" + "=" * 70)
print(f"Thresholds used -> High volume: >= {vol_threshold:.1f} reviews | "
      f"Low satisfaction: <= {rating_threshold:.2f} avg rating")
print("=" * 70)

risk_districts = district_summary[
    (district_summary["review_volume"] >= vol_threshold) &
    (district_summary["avg_rating"] <= rating_threshold)
    ].sort_values(["review_volume", "avg_rating"], ascending=[False, True])

# Key Findings
if len(risk_districts) > 0:
    print("Key Findings")
    print("=" * 70)
    print("\nDISTRICTS THAT ARE BOTH HIGH-VOLUME AND LOW-SATISFACTION:")
    print(risk_districts.to_string(index=False))
else:
    print("\nNo district is simultaneously in the top-volume AND bottom-rating quartile.")
    print("(High volume and low satisfaction don't overlap strongly in this data.)")

# STATISTICAL OUTLIERS (IQR method) on avg_rating
q1, q3 = district_summary["avg_rating"].quantile([0.25, 0.75])
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

rating_outliers = district_summary[
    (district_summary["avg_rating"] < lower_bound) |
    (district_summary["avg_rating"] > upper_bound)
    ].sort_values("avg_rating")

print("\n" + "=" * 70)
print(f"STATISTICAL OUTLIERS on avg_rating (IQR method, bounds: "
      f"{lower_bound:.2f} to {upper_bound:.2f})")
print("=" * 70)
if len(rating_outliers) > 0:
    print(rating_outliers.to_string(index=False))
else:
    print("No statistical outliers detected on avg_rating using the IQR method.")

# Flag low-confidence outliers (very small sample size)
LOW_SAMPLE_THRESHOLD = 3
low_confidence = rating_outliers[rating_outliers["review_volume"] <= LOW_SAMPLE_THRESHOLD]
if len(low_confidence) > 0:
    print(f"\nCAUTION: {len(low_confidence)} of these outlier district(s) have "
          f"<= {LOW_SAMPLE_THRESHOLD} reviews — their rating could swing heavily "
          f"with just one more review. Treat with care.")

# ============================================================
# Plotly Visualizations
# ============================================================
print("\n" + "=" * 70)
print("RENDERING VISUALIZATIONS")
print("=" * 70)

# Display Chart 1: Top 10 districts by review volume

top10_volume = district_summary.sort_values("review_volume", ascending=False).head(10).copy()
top10_volume["district_label"] = top10_volume["district_id"].astype(str)

fig1 = px.bar(
    top10_volume,
    x="review_volume",
    y="district_label",
    orientation="h",
    color="review_volume",
    color_continuous_scale="Blues",
    text="review_volume",
    title="Top 10 Districts by Review Volume",
    labels={"review_volume": "Review Volume", "district_label": "District"},
)
fig1.update_traces(textposition="outside")
fig1.update_layout(
    template="plotly_white",
    yaxis=dict(autorange="reversed"),
    coloraxis_showscale=False,
)
fig1.show()

# Display Chart 2: Bottom 10 districts by average rating

bottom10_rating = district_summary.sort_values("avg_rating", ascending=True).head(10).copy()
bottom10_rating["district_label"] = bottom10_rating["district_id"].astype(str)

fig2 = px.bar(
    bottom10_rating,
    x="avg_rating",
    y="district_label",
    orientation="h",
    color="avg_rating",
    color_continuous_scale="Reds_r",
    text="avg_rating",
    title="Bottom 10 Districts by Average Rating (Lowest Satisfaction)",
    labels={"avg_rating": "Average Rating", "district_label": "District"},
)
fig2.update_traces(textposition="outside")
fig2.update_layout(
    template="plotly_white",
    yaxis=dict(autorange="reversed"),
    coloraxis_showscale=False,
    xaxis_range=[0, 5],
)
fig2.show()

# Display Chart 3: Risk quadrant-volume vs. rating, with outliers flagged

risk_ids = set(risk_districts["district_id"])
outlier_ids = set(rating_outliers["district_id"])

def classify_district(row):
    in_risk = row["district_id"] in risk_ids
    in_outlier = row["district_id"] in outlier_ids
    if in_risk and in_outlier:
        return "High Volume + Low Satisfaction & Statistical Outlier"
    elif in_risk:
        return "High Volume + Low Satisfaction"
    elif in_outlier:
        return "Statistical Outlier (IQR)"
    else:
        return "Normal"

quadrant_df = district_summary.copy()
quadrant_df["district_label"] = quadrant_df["district_id"].astype(str)
quadrant_df["risk_flag"] = quadrant_df.apply(classify_district, axis=1)

color_map = {
    "Normal": "#94a3b8",
    "High Volume + Low Satisfaction": "#dc2626",
    "Statistical Outlier (IQR)": "#f59e0b",
    "High Volume + Low Satisfaction & Statistical Outlier": "#7c1d1d",
}

fig3 = px.scatter(
    quadrant_df,
    x="review_volume",
    y="avg_rating",
    color="risk_flag",
    color_discrete_map=color_map,
    size="pct_1_star",
    size_max=20,
    hover_name="district_label",
    hover_data={"pct_1_star": True, "district_label": False},
    labels={"review_volume": "Review Volume", "avg_rating": "Average Rating"},
)

# Threshold lines (no inline text — labels are placed in the margins instead,
# so they never sit on top of a data point)
fig3.add_vline(x=vol_threshold, line_dash="dash", line_color="#94a3b8", line_width=1.5)
fig3.add_hline(y=rating_threshold, line_dash="dash", line_color="#94a3b8", line_width=1.5)

fig3.add_annotation(
    x=vol_threshold, y=1.06, xref="x", yref="paper",
    text=f"High-volume threshold ({vol_threshold:.0f} reviews)",
    showarrow=False, font=dict(size=11, color="#64748b"),
)
fig3.add_annotation(
    x=1.02, y=rating_threshold, xref="paper", yref="y",
    text=f"Low-satisfaction<br>threshold ({rating_threshold:.2f})",
    showarrow=False, font=dict(size=11, color="#64748b"),
    xanchor="left", align="left",
)

fig3.update_layout(
    title=dict(
        text="District Risk Quadrant: Review Volume vs. Average Rating",
        x=0.02, xanchor="left",
    ),
    template="plotly_white",
    legend=dict(
        orientation="h",
        yanchor="top", y=-0.22,
        xanchor="center", x=0.5,
        title_text="",
    ),
    margin=dict(t=90, r=160, b=120, l=70),
    height=580,
)
fig3.update_yaxes(range=[0, 5.3])
fig3.show()
```

```python
# ==========================================
# Page 04
# CRM & Customer Experience
#
# Theme 5: Geography,Trends & Behavior
# Question 14
# Is complaint volume rising, falling, or stable over time, and has the bank-vs-credit-card mix shifted?
# Aparupa Chandra
# ==========================================
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Load Dataset
EVENTS_PATH = "crm_events.csv"

events = pd.read_csv(EVENTS_PATH)
events["date_received"] = pd.to_datetime(events["date_received"], format="%d-%m-%Y")
events["year"] = events["date_received"].dt.year
events["year_month"] = events["date_received"].dt.to_period("M")

print(f"Data range: {events['date_received'].min().date()} to {events['date_received'].max().date()}")
print(f"Total complaints: {len(events):,}\n")

# YEARLY COMPLAINT VOLUME TREND
yearly_volume = events.groupby("year").size().reset_index(name="complaint_count")

# NOTE: first and last years may be partial (data doesn't start/end on Jan 1 / Dec 31)
yearly_volume["pct_change_vs_prev_year"] = yearly_volume["complaint_count"].pct_change().mul(100).round(1)

# Display Result
print("\n"+"=" * 70)
print("Question 14 Result")
print("=" * 70)
print("YEARLY COMPLAINT VOLUME")
print("=" * 70)
print(yearly_volume.to_string(index=False))
print("\nNOTE: 2011 and 2017 are likely PARTIAL years (data starts Dec 2011, ends Apr 2017)")
print("      -exclude them when judging the real year-over-year trend.")

# MONTHLY TREND (finer-grained, full series)
monthly_volume = events.groupby("year_month").size().reset_index(name="complaint_count")
monthly_volume["year_month"] = monthly_volume["year_month"].astype(str)

print("\n" + "=" * 70)
print("FIRST 6 AND LAST 6 MONTHS (full monthly series saved to CSV)")
print("=" * 70)
print("First 6 months:")
print(monthly_volume.head(6).to_string(index=False))
print("\nLast 6 months:")
print(monthly_volume.tail(6).to_string(index=False))

# SIMPLE TREND DIRECTION using full years only (2012-2016)
full_years = yearly_volume[(yearly_volume["year"] >= 2012) & (yearly_volume["year"] <= 2016)]

# Key Findings For Complaint Review
print("\n" + "=" * 70)
print("Key Findings For Complaint Volume Rising, Falling, or Stable over time")
print("=" * 70)
print("TREND DIRECTION-based on complete years only (2012-2016)")
print("=" * 70)
print(full_years.to_string(index=False))

# Linear trend via simple slope check
if len(full_years) >= 2:
    x = full_years["year"].values
    y = full_years["complaint_count"].values
    slope = ((x - x.mean()) * (y - y.mean())).sum() / ((x - x.mean()) ** 2).sum()
    direction = "RISING" if slope > 0 else "FALLING" if slope < 0 else "STABLE"
    print(f"\nApprox. linear trend slope: {slope:.1f} complaints/year -> Overall direction: {direction}")

# BANK vs CREDIT CARD MIX — has the split shifted over time?
mix_by_year = (
    events.groupby(["year", "product"]).size().unstack(fill_value=0)
)
mix_pct_by_year = mix_by_year.div(mix_by_year.sum(axis=1), axis=0).mul(100).round(1)

print("\n" + "=" * 70)
print("BANK vs CREDIT CARD COMPLAINT COUNT BY YEAR")
print("=" * 70)
print(mix_by_year.to_string())

print("\n" + "=" * 70)
print("BANK vs CREDIT CARD MIX (% of that year's complaints)")
print("=" * 70)
print(mix_pct_by_year.to_string())

# Quantify the shift: compare first full year vs last full year
if "Bank account or service" in mix_pct_by_year.columns:
    first_yr, last_yr = full_years["year"].min(), full_years["year"].max()
    bank_start = mix_pct_by_year.loc[first_yr, "Bank account or service"]
    bank_end = mix_pct_by_year.loc[last_yr, "Bank account or service"]
    print("\nKey Findings For BANK vs CREDIT Card MiX ")
    print("=" * 70)
    print(f"'Bank account or service' share: {bank_start:.1f}% in {first_yr} -> {bank_end:.1f}% in {last_yr}")
    shift = bank_end - bank_start
    if abs(shift) < 3:
        print(f"Mix has stayed roughly STABLE (change of {shift:+.1f} percentage points).")
    else:
        direction_word = "toward Bank account/service" if shift > 0 else "toward Credit card"
        print(f"Mix HAS SHIFTED by {shift:+.1f} percentage points, {direction_word}.")

# ============================================================
# Plotly Visualizations
# ============================================================
print("\n" + "=" * 70)
print("RENDERING VISUALIZATIONS")
print("=" * 70)

# Display Chart 1: Monthly complaint volume over time (full series)

monthly_plot_df = monthly_volume.copy()
monthly_plot_df["year_month_dt"] = pd.PeriodIndex(monthly_plot_df["year_month"], freq="M").to_timestamp()

fig1 = px.line(
    monthly_plot_df,
    x="year_month_dt",
    y="complaint_count",
    markers=True,
    labels={"year_month_dt": "Month", "complaint_count": "Complaint Count"},
)
fig1.update_traces(line_color="#2563eb", marker=dict(size=5), name="Monthly complaints", showlegend=True)

if len(full_years) > 0:
    full_year_start = pd.Timestamp(f"{int(full_years['year'].min())}-01-01")
    full_year_end = pd.Timestamp(f"{int(full_years['year'].max())}-12-31")

    # Shade the partial years at each end (per the note above) instead of
    # labeling them inline, so the label never collides with the data
    fig1.add_vrect(
        x0=monthly_plot_df["year_month_dt"].min(), x1=full_year_start,
        fillcolor="#94a3b8", opacity=0.18, line_width=0,
    )
    fig1.add_vrect(
        x0=full_year_end, x1=monthly_plot_df["year_month_dt"].max(),
        fillcolor="#94a3b8", opacity=0.18, line_width=0,
    )
    # Invisible trace purely to give the shaded region a legend entry
    fig1.add_trace(go.Scatter(
        x=[None], y=[None], mode="markers",
        marker=dict(size=10, color="#94a3b8", opacity=0.5, symbol="square"),
        name="Partial year (excluded from trend)",
    ))

fig1.update_layout(
    title=dict(text="Monthly Complaint Volume Over Time", x=0.02, xanchor="left"),
    template="plotly_white",
    legend=dict(orientation="h", yanchor="bottom", y=-0.3, xanchor="center", x=0.5, title_text=""),
    margin=dict(t=60, b=110),
    height=480,
)
fig1.show()

# Display Chart 2: Yearly complaint volume, full vs partial years, with trend line

yearly_plot_df = yearly_volume.copy()
yearly_plot_df["period_type"] = yearly_plot_df["year"].apply(
    lambda yr: "Full year" if yr in full_years["year"].values else "Partial year"
)

fig2 = px.bar(
    yearly_plot_df,
    x="year",
    y="complaint_count",
    color="period_type",
    color_discrete_map={"Full year": "#2563eb", "Partial year": "#cbd5e1"},
    text="complaint_count",
    labels={"complaint_count": "Complaint Count", "year": "Year"},
)
fig2.update_traces(textposition="outside")

if len(full_years) >= 2:
    x_full = full_years["year"].values
    y_full = full_years["complaint_count"].values
    slope_viz = ((x_full - x_full.mean()) * (y_full - y_full.mean())).sum() / ((x_full - x_full.mean()) ** 2).sum()
    intercept_viz = y_full.mean() - slope_viz * x_full.mean()
    trend_y = slope_viz * x_full + intercept_viz
    fig2.add_trace(go.Scatter(
        x=x_full, y=trend_y, mode="lines", name="Trend (full years)",
        line=dict(color="#dc2626", dash="dash", width=2),
    ))

fig2.update_layout(
    title=dict(text="Yearly Complaint Volume (partial years shaded gray)", x=0.02, xanchor="left"),
    template="plotly_white",
    legend=dict(orientation="h", yanchor="bottom", y=-0.3, xanchor="center", x=0.5, title_text=""),
    margin=dict(t=60, b=110),
)
fig2.update_xaxes(dtick=1)
fig2.show()

# Display Chart 3: Bank vs. credit card complaint mix by year

mix_pct_long = mix_pct_by_year.reset_index().melt(id_vars="year", var_name="product", value_name="pct")

fig3 = px.bar(
    mix_pct_long,
    x="year",
    y="pct",
    color="product",
    text="pct",
    labels={"pct": "% of Year's Complaints", "year": "Year", "product": "Product"},
)
fig3.update_traces(texttemplate="%{text:.1f}%", textposition="inside")
fig3.update_layout(
    title=dict(text="Bank vs. Credit Card Complaint Mix by Year", x=0.02, xanchor="left"),
    template="plotly_white",
    barmode="stack",
    yaxis_range=[0, 100],
    legend=dict(orientation="h", yanchor="bottom", y=-0.3, xanchor="center", x=0.5, title_text=""),
    margin=dict(t=60, b=110),
)
fig3.update_xaxes(dtick=1)
fig3.show()
```

```python
# ==========================================
# Page 04
# CRM & Customer Experience
#
# Theme 5: Geography,Trends & Behavior
# Question 15
# Do complaints come from a broad customer base or a small group of repeat complainers, and are they treated differently?
# Aparupa Chandra
# ==========================================

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Load Dataset
EVENTS_PATH = "crm_events.csv"
events = pd.read_csv(EVENTS_PATH)

RESOLVED_STATUSES = [
    "Closed with relief",
    "Closed with monetary relief",
    "Closed with non-monetary relief",
]
events["is_resolved"] = events["company_response_to_consumer"].isin(RESOLVED_STATUSES)
events["is_timely"] = events["timely_response"].str.strip().str.lower() == "yes"

# BROAD BASE vs CONCENTRATED — basic counts
complaints_per_client = events["client_id"].value_counts()

total_complaints = len(events)
total_clients = events["client_id"].nunique()
one_time = (complaints_per_client == 1).sum()
repeat = (complaints_per_client >= 2).sum()

# Display Result
print("\n" + "=" * 70)
print("Question 15 Result")
print("=" * 70)
print("CUSTOMER BASE OVERVIEW")
print("=" * 70)
print(f"Total complaints       : {total_complaints:,}")
print(f"Unique clients         : {total_clients:,}")
print(f"Avg complaints/client   : {complaints_per_client.mean():.2f}")
print(f"Median complaints/client: {complaints_per_client.median():.0f}")
print(f"One-time complainers    : {one_time:,} ({one_time/total_clients*100:.1f}% of clients)")
print(f"Repeat complainers (2+) : {repeat:,} ({repeat/total_clients*100:.1f}% of clients)")
print(f"Max complaints by 1 client: {complaints_per_client.max()}")

# CONCENTRATION (Pareto-style) — how much volume comes from
#    the heaviest-complaining customers?
sorted_counts = complaints_per_client.sort_values(ascending=False)
cumulative_pct = sorted_counts.cumsum() / total_complaints * 100

print("\n" + "=" * 70)
print("COMPLAINT CONCENTRATION (Pareto check)")
print("=" * 70)
for pct in [0.10, 0.20, 0.50]:
    n = int(total_clients * pct)
    share = sorted_counts.iloc[:n].sum() / total_complaints * 100
    print(f"Top {int(pct*100)}% of clients ({n:,} clients) generate {share:.1f}% of all complaints")

# DISTRIBUTION OF COMPLAINT COUNTS PER CLIENT
dist = complaints_per_client.value_counts().sort_index()
print("\n" + "=" * 70)
print("DISTRIBUTION: number of clients by complaint count")
print("=" * 70)
print(dist.to_string())

# ARE REPEAT COMPLAINERS TREATED DIFFERENTLY?
#    Compare resolution rate & timely response: one-time vs repeat
events["complaint_count_for_client"] = events["client_id"].map(complaints_per_client)
events["complainer_type"] = pd.cut(
    events["complaint_count_for_client"],
    bins=[0, 1, 3, 6, 100],
    labels=["1 (one-time)", "2-3", "4-6", "7+"]
)

treatment_by_type = (
    events.groupby("complainer_type", observed=False)
    .agg(
        complaint_count=("complaint_id", "count"),
        unique_clients=("client_id", "nunique"),
        resolution_rate=("is_resolved", "mean"),
        timely_response_rate=("is_timely", "mean"),
    )
    .reset_index()
)
treatment_by_type["resolution_rate"] = (treatment_by_type["resolution_rate"] * 100).round(2)
treatment_by_type["timely_response_rate"] = (treatment_by_type["timely_response_rate"] * 100).round(2)

print("\n" + "=" * 70)
print("TREATMENT COMPARISON: one-time vs repeat complainers")
print("=" * 70)
print(treatment_by_type.to_string(index=False))

# Quick comparison: one-time vs all repeat (2+)
one_time_mask = events["complaint_count_for_client"] == 1
repeat_mask = events["complaint_count_for_client"] >= 2

#Key Findings
print("\nKey Findings For Treatment for One Time Complaint & Repeat Complaint Customers")
print("=" * 70)
print("SIMPLE ONE-TIME vs REPEAT (2+) COMPARISON")
print("=" * 70)
print(f"One-time complainers -> resolution rate: {events.loc[one_time_mask, 'is_resolved'].mean()*100:.2f}%, "
      f"timely response: {events.loc[one_time_mask, 'is_timely'].mean()*100:.2f}%")
print(f"Repeat complainers    -> resolution rate: {events.loc[repeat_mask, 'is_resolved'].mean()*100:.2f}%, "
      f"timely response: {events.loc[repeat_mask, 'is_timely'].mean()*100:.2f}%")

res_diff = events.loc[repeat_mask, 'is_resolved'].mean()*100 - events.loc[one_time_mask, 'is_resolved'].mean()*100
print(f"\nResolution rate gap (repeat - one-time): {res_diff:+.2f} percentage points")
if abs(res_diff) < 2:
    print("-> No meaningful difference in treatment by complainer type.")
elif res_diff > 0:
    print("-> Repeat complainers get resolved MORE often than one-time complainers.")
else:
    print("-> Repeat complainers get resolved LESS often than one-time complainers.")

# ============================================================
# Plotly Visualizations
# ============================================================
print("\n" + "=" * 70)
print("RENDERING VISUALIZATIONS")
print("=" * 70)

# Display Chart 1: Complaint concentration (Lorenz-style curve)

pareto_df = pd.DataFrame({
    "client_rank": range(1, len(sorted_counts) + 1),
    "cumulative_complaint_pct": cumulative_pct.values,
})
pareto_df["cumulative_client_pct"] = pareto_df["client_rank"] / len(sorted_counts) * 100

# Recomputing  the same three checkpoints printed above, just to plot as markers
pareto_points = []
for pct in [0.10, 0.20, 0.50]:
    n = int(total_clients * pct)
    share = sorted_counts.iloc[:n].sum() / total_complaints * 100
    pareto_points.append((pct * 100, share))

fig1 = go.Figure()
fig1.add_trace(go.Scatter(
    x=[0, 100], y=[0, 100], mode="lines",
    line=dict(color="#cbd5e1", dash="dash"), name="Equal distribution (reference)",
))
fig1.add_trace(go.Scatter(
    x=pareto_df["cumulative_client_pct"], y=pareto_df["cumulative_complaint_pct"],
    mode="lines", name="Actual complaint concentration",
    line=dict(color="#dc2626", width=3),
))
for pct, share in pareto_points:
    fig1.add_trace(go.Scatter(
        x=[pct], y=[share], mode="markers+text",
        marker=dict(size=9, color="#1e3a8a"),
        text=[f"{share:.0f}%"], textposition="top center",
        showlegend=False,
    ))

fig1.update_layout(
    title=dict(text="Complaint Concentration: Top Clients vs. Share of Complaints", x=0.02, xanchor="left"),
    xaxis_title="Cumulative % of Clients (ranked by complaint count)",
    yaxis_title="Cumulative % of Complaints",
    template="plotly_white",
    legend=dict(orientation="h", yanchor="bottom", y=-0.3, xanchor="center", x=0.5, title_text=""),
    margin=dict(t=60, b=110),
    xaxis_range=[0, 100],
    yaxis_range=[0, 100],
)
fig1.show()

# Display Chart 2: Distribution of clients by complaint count

dist_df = dist.rename_axis("complaints_per_client").reset_index(name="num_clients")

fig2 = px.bar(
    dist_df,
    x="complaints_per_client",
    y="num_clients",
    text="num_clients",
    color_discrete_sequence=["#2563eb"],
    labels={"complaints_per_client": "Complaints per Client", "num_clients": "Number of Clients"},
    log_y=True,
)
fig2.update_traces(textposition="outside")
fig2.update_layout(
    title=dict(text="Distribution of Clients by Complaint Count (log scale)", x=0.02, xanchor="left"),
    template="plotly_white",
    margin=dict(t=60),
)
fig2.update_xaxes(dtick=1)
fig2.show()

# Display Chart 3: Treatment comparison by complainer type

treatment_long = treatment_by_type.melt(
    id_vars=["complainer_type", "complaint_count", "unique_clients"],
    value_vars=["resolution_rate", "timely_response_rate"],
    var_name="metric", value_name="rate",
)
metric_labels = {
    "resolution_rate": "Resolution Rate (%)",
    "timely_response_rate": "Timely Response Rate (%)",
}
treatment_long["metric"] = treatment_long["metric"].map(metric_labels)

fig3 = px.bar(
    treatment_long,
    x="complainer_type",
    y="rate",
    color="metric",
    barmode="group",
    text="rate",
    category_orders={"complainer_type": ["1 (one-time)", "2-3", "4-6", "7+"]},
    color_discrete_map={
        "Resolution Rate (%)": "#2563eb",
        "Timely Response Rate (%)": "#f59e0b",
    },
    labels={"complainer_type": "Complainer Type", "rate": "Rate (%)"},
)
fig3.update_traces(texttemplate="%{text:.1f}%", textposition="outside")
fig3.update_layout(
    title=dict(text="Treatment Comparison: One-Time vs. Repeat Complainers", x=0.02, xanchor="left"),
    template="plotly_white",
    yaxis_range=[0, 100],
    legend=dict(orientation="h", yanchor="bottom", y=-0.3, xanchor="center", x=0.5, title_text=""),
    margin=dict(t=60, b=110),
)
fig3.show()
```
